as mentioned, labels in main dataset on kaggle are like for example CP_XXX_XXX_.these labeles need to be mapped accordingly to standard format . this is also something that could be done obviously in the main code, but in order to avoid messing up with the original code , here it is done seperatly

```python
import os
from tqdm import tqdm
import time
import shutil

tic = time.perf_counter()
files = os.listdir('path')
toc = time.perf_counter()
print("files read in ",toc-tic," seconds")

tic = time.perf_counter()
for img in tqdm(files):
    name=img.split('_')
    if name[0]=='NCP':
        name[0]='2'
    elif name[0]=='CP':
        name[0]='1'
    elif name[0]=='Normal':
        name[0]='0'
    new_name = '_'.join(name)
    os.rename('path/'+img,'path/'+new_name)
toc = time.perf_counter()
print("rename done in ",(toc-tic)/60," minutes")

```
